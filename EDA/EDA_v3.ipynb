### 1. \n, ...., ^, 제거
### 2. 구두점 제거
### 3. 이모티콘 제거
### 4. 불용어 제거

```python
def preprocess(review):
    review = re.sub(r'\.{2,}', '', review)  # ........있는 부분 제거
    review = re.sub(r'[\n\r\t]', ' ', review)  # escape sequence 제거
    review = review.lower()  # 소문자로 변환
    # ---Rated via Beer Buddy for iPhone 부분 잘라내기
    review = review.split('---rated')[0]
    return review
```

```python
import re

df['Review'] = df['Review'].apply(preprocess)
df.reset_index(drop=True, inplace=True)
```

```python
df['Review'][0]
```

```python
df['Review'] = df['Review'].str.replace('[^\w\s&]', '')
df['Review'][0]
```

```python
# 이모티콘 제거

def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags 
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
df['Review'] = df['Review'].apply(lambda x: remove_emoji(x))
```

```python
# 불용어 제거

import nltk
from nltk.corpus import stopwords
import spacy

nltk.download('stopwords')
```

### VADER를 사용하여 감정 분류수행하기

```python
import nltk
from tqdm import tqdm

nltk.download('vader_lexicon')

# vader불러오기

from nltk.sentiment.vader import SentimentIntensityAnalyzer
vds = SentimentIntensityAnalyzer()

# 감정별 컬럼 생성하기
df = df.assign(neg = '', neu = '', pos = '', compound = '')
df

# 모든 데이터에 적용하기

def vader(review):
    sen = vds.polarity_scores(review)
    return sen

# 감정 유형별 score 구하기
sentiment_neg = []
sentiment_neu = []
sentiment_pos = []
sentiment_compound = []

for s in tqdm(df['Review']):
    multi_sen = vader(s)
    sentiment_neg.append(multi_sen['neg'])
    sentiment_neu.append(multi_sen['neu'])
    sentiment_pos.append(multi_sen['pos'])
    sentiment_compound.append(multi_sen['compound'])
    
    
df['neg'] = sentiment_neg
df['neu'] = sentiment_neu
df['pos'] = sentiment_pos
df['compound'] = sentiment_compound
```

```python
df
```

```python
df['result'] = ''

for i in tqdm(range(len(df))):
    if df['compound'][i] >= 0.05:
        df['result'][i] = 'positive'
    elif df['compound'][i] <= -0.05:
        df['result'][i] = 'negative'
    else:
        df['result'][i] = 'neutral'
```

```python
# Review 데이터를 positive, neutral, negative로 라벨링하는 함수
def label_review(review):
    prompt = f"Please classify the following review as positive, neutral, or negative.\nReview: {review}\nSentiment:"
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=1,
        n=1,
        stop=None,
        temperature=0.7,
    )
    sentiment = response.choices[0].text.strip().lower()
    if sentiment == "positive":
        return "positive"
    elif sentiment == "negative":
        return "negative"
    else:
        return "neutral"
```

```python
from tqdm.notebook import tqdm

df_test['Label'] = tqdm(df_test['Review'].apply(label_review), desc="Labeling Reviews")
```
